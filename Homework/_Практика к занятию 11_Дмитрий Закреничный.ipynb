### Оценим распределение выборочного среднего для разных объёмов выборок
Для этого постройте на одном полотне гистограмму выборочных средних и теоретическое распределение выборочных средних по ЦПТ для разных объёмов выборок.

Для теоретического распределение используйте настоящие значения мат ожидания и дисперсии (не приближенные на основе выборки).

```python
plt.figure(figsize=(25, 15))
size_sample = [10, 30, 50, 100]
q_means = [100, 500, 1000, 10000]
for ind, val in enumerate(size_sample):
  distribution_means = sts.norm(loc=laplace_rv.mean(), scale=laplace_rv.std() / np.sqrt(val))
  for i in range(4):
    plt.subplot(4, 4, ind+1 + 4*i)
    means = []
    for q in range(q_means[i]):
      means.append(np.mean(laplace_rv.rvs(val)))

    plt.hist(means, bins='auto', density=True)

    left, right = min(means), max(means)
    x = np.linspace(left, right, 100)
    means_pdf = distribution_means.pdf(x)
    plt.plot(x, means_pdf)
    
    plt.title('{} элементов {} выборок'.format(val, q_means[i]))
    plt.legend(['CLT means', 'real means'])
```

## Доверительный интервал на основе ЦПТ
Возьмите данные для оттока телеком компании, и постройте доверительные интервалы для среднего значения total day calls, используя центральную предельную теорему в разбиении по колонке churn (ушел/не ушел). 

Для оценки дисперсии используйте несмещенную оценку (параметр ddof=1).

Сделайте оценку с использованием явной формулы для z интервала, а затем с помощью функций 
`from statsmodels.stats.weightstats import _zconfint_generic`. Сравните результаты.


Ссылка на данные: https://drive.google.com/file/d/1X_drKqoarnU2Sz2har7uSCsgyhh8Xtz8/view?usp=sharing

```python
df = pd.read_csv('telecom_churn.csv')
df.head(5)
```

```python
df_day_calls = df[['total day calls', 'churn']]
df_day_calls.head(5)
```

```python
df_day_calls.groupby('churn').size()
```

```python
df_day_calls.groupby('churn')['total day calls'].mean()
```

```python
churn = df_day_calls[df_day_calls['churn'] == True]['total day calls']
no_churn = df_day_calls[df_day_calls['churn'] == False]['total day calls']
```

```python
from statsmodels.stats.weightstats import _zconfint_generic
```

```python
alpha = 0.05
standart_rv = sts.norm(0, 1)
dzeta = standart_rv.ppf(1 - alpha/2)

churn_mean = churn.mean()
churn_std = churn.std(ddof=1)
no_churn_mean = no_churn.mean()
no_churn_std = no_churn.std(ddof=1)

churn_a = churn_mean - dzeta*churn_std/np.sqrt(len(churn))
churn_b = churn_mean + dzeta*churn_std/np.sqrt(len(churn))
no_churn_a = no_churn_mean - dzeta*no_churn_std/np.sqrt(len(no_churn))
no_churn_b = no_churn_mean + dzeta*no_churn_std/np.sqrt(len(no_churn))

print('result with explicit formula z-interval')
print('ME "total day calls" for churn is {:.2f} and {:.2f}'.format(churn_a, churn_b))
print('ME "total day calls" for no_churn is {:.2f} and {:.2f}'.format(no_churn_a, no_churn_b))

print('\nresult with "_zconfint_generic"')
print(_zconfint_generic(churn_mean, churn_std/np.sqrt(len(churn)), 0.05, 'two-sided'), 'for churn')
print(_zconfint_generic(no_churn_mean, no_churn_std/np.sqrt(len(no_churn)), 0.05, 'two-sided'), 'for no_churn')
```
