### Задание 33
Выбрать названия отдела, заключившего сделки на самую большую сумму, и города, в котором он находится. Здесь удобно применить подзапросы, хотя можно сделать и с помощью соединений таблиц.

### Задание 34
Выбрать название товара, дату продажи, цену продажи для всех случаев, когда товары продавались ниже, чем за 75% их объявленной цены.

**В заданиях подобным этому, не забудьте о том как действует цена в этой базе: она действует от price.start_date до price.end_date, причем если price.end_date IS NULL, то цена действует до сих пор!**

```
pd.read_sql_query("""
SELECT p.description,
       s_o.order_date,
       (i.actual_price / pr.list_price * 100) AS per
FROM price AS pr
JOIN product AS p ON p.product_id = pr.product_id
JOIN item AS i ON i.product_id = p.product_id
JOIN sales_order AS s_o ON s_o.order_id = i.order_id
WHERE ((s_o.order_date BETWEEN pr.start_date AND pr.end_date)
       OR (s_o.order_date > pr.start_date
           AND pr.end_date IS NULL))
  AND (i.actual_price / pr.list_price * 100 < 75)
""", conn)
```

### Задание 35
Выбрать общую сумму скидок, предоставленных покупателю STADIUM SPORTS в 1989г.

**Скидка здесь это list_price*quantity-item.total**

```
pd.read_sql_query("""
SELECT sum(pr.list_price * i.quantity - i.total) AS sum_disc
FROM price AS pr
JOIN product AS p ON p.product_id = pr.product_id
JOIN item AS i ON i.product_id = p.product_id
JOIN sales_order AS s_o ON s_o.order_id = i.order_id
JOIN customer AS c ON s_o.customer_id = c.customer_id
WHERE (c.name = 'stadium sports')
  AND (date_part('year', s_o.order_date) = 1989)
  AND ((s_o.order_date BETWEEN pr.start_date AND pr.end_date)
       OR (s_o.order_date > pr.start_date
           AND pr.end_date IS NULL))
""", conn)
```

### Задание 40
Выбрать названия товаров, для которых нынешняя цена увеличилась по сравнению с ценой на 15 декабря 1989г. более, чем на 15%.

Здесь удобно делать задачу по частям

1. Получите цену на 15 декабря 1989г.
2. Получите цену, действующую сейчас
3. Соедините эти два запроса

Сравнивать тут нужно только list_price.

```
pd.read_sql_query("""
SELECT description
FROM product
WHERE product_id in
    (SELECT price_now.product_id
     FROM
       (SELECT product_id,
               list_price
        FROM price
        WHERE end_date IS NULL)AS price_now
     JOIN
       (SELECT product_id,
               list_price
        FROM price
        WHERE (start_date <= '1989-12-15')
          AND (end_date >= '1989-12-15'
               OR end_date IS NULL)) AS price_old ON price_now.product_id = price_old.product_id
     WHERE price_now.list_price > price_old.list_price)

""", conn)
```
